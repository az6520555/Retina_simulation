```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import signal
from scipy.fftpack import fft, fftfreq
from scipy.io import loadmat
import xlsxwriter 
%matplotlib inline
```

```python
path='F:/我的雲端硬碟/Retina exp/exp data/Sorted_final_data/20200419/MIandSTA/'
file=['OU_tau=1000ms_19-Apr-2020_0_sort_unit1_STA','OU_tau=1000ms_cutoff=10_19-Apr-2020_0_sort_unit1_STA',\
      'OU_tau=1000ms_cutoff=7_19-Apr-2020_0_sort_unit1_STA','OU_tau=1000ms_cutoff=4_19-Apr-2020_0_sort_unit1_STA',\
     'OU_tau=1000ms_cutoff=2_19-Apr-2020_0_sort_unit1_STA']

select_chan=39
STA=[[] for i in range(len(file))]
for i in range(len(file)):
    dat=loadmat(path+file[i]+'.mat')
    STA[i]=dat['STAAAAA'][0][select_chan-1][0]
    Timelag=dat['TimeShift'][0][select_chan-1][0]
    
```

```python
colorlist=['#1f77b4','#ff7f0e','#d62728','#9467bd','#2ca02c']
for i in range(len(file)):
    plt.plot(Timelag,STA[i],color=colorlist[i])
```

```python
# save excel file of STA data
workbook = xlsxwriter.Workbook('F:/我的雲端硬碟/Retina exp/Retina as NGD paper submit/data/0419 STA data channel 39.xlsx') 
worksheet = workbook.add_worksheet()
for i in range(len(Timelag)):
    worksheet.write(i,0,Timelag[i])
    for j in range(len(STA)):
        worksheet.write(i,j+1,STA[j][i])
workbook.close()
```
