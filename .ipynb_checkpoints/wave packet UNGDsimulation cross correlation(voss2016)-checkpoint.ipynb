```python
# wave packet
omega1=0.35*2*math.pi
omega2=0.4*2*math.pi
T=np.linspace(0,100,10000)
dt=T[1]-T[0]
xa=np.zeros(len(T))
xb=np.zeros(len(T))
x=np.zeros(len(T))
for i in range(len(T)):
    xa[i]=math.sin(omega1*T[i])
    xb[i]=math.sin(omega2*T[i])
    x[i]=xa[i]+xb[i]
plt.plot(T,x)
plt.xlim()
```

```python
# UNGD output
tau=2
alpha=5.3
K=5.4
y=np.zeros(len(T))
for i in range(len(T)-1):
    delayi=int(i-tau/dt)
    if delayi>=0:
        dy=dt*(-alpha*y[i]+K*(x[i]-y[int(delayi)]))
    else:
        dy=0
    y[i+1]=y[i]+dy
plt.plot(T,x)
plt.plot(T,y/max(y)*max(x))
plt.xlim(40,80)
```
