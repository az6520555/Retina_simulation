```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
# alpha = 20
# beta =  1

# K = 20
# g = 10
alpha, beta, K, g =8,1,40,10
```

```python
omega=np.arange(0,500,0.01)
A=np.zeros(len(omega))
B=np.zeros(len(omega))
for i in range(len(omega)):
    A[i]=alpha*beta**2+g*beta*K+omega[i]**2*alpha
    B[i]=-omega[i]**3+omega[i]*g*K-beta**2*omega[i]
plt.plot(A)
plt.plot(B)
plt.show()
# plt.plot(B/A)
plt.show()
```

```python
# phase
phi=np.zeros(len(omega))
for i in range(len(omega)):
    phi[i]=math.atan(B[i]/A[i])
plt.plot(omega/2/math.pi,phi)
```

```python
# group
delta=-np.diff(phi)/(omega[1]-omega[0])
plt.plot(omega[0:-1]/2/math.pi,delta)
plt.ylim()
```

```python
# gain
C=np.zeros(len(omega))
for i in range(len(omega)):
    C[i]=(-omega[i]**2+alpha*beta+g*K)**2+(omega[i]*(alpha+beta))**2
gain=K/C*(A**2+B**2)**(1/2)
# plt.plot(C)
plt.show()
plt.plot(omega/2/math.pi,gain)
```

```python
# plot together
plt.plot(omega/2/math.pi,phi)
plt.plot(omega[0:-1]/2/math.pi,delta)
plt.plot(omega/2/math.pi,gain)
plt.legend(('phi','group delay','gain'))
plt.xlim()
plt.grid()
```
