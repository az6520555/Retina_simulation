```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
tau=2
K=30
alpha=10
omega=np.linspace(0,50,1000)
beta=np.zeros(len(omega))
for i in range(len(omega)):
    beta[i]=(alpha+K*math.cos(omega[i]*tau))**2+(omega[i]-K*math.sin(omega[i]*tau))**2
G=K*(np.ones(len(beta))/beta)
plt.plot(omega/2/math.pi,G)
# plt.xlim(0,1)
# plt.ylim(0,1)
plt.show()
```

```python
tanphi=np.zeros(len(omega))
phi=np.zeros(len(omega))
sign=1
B=0
for i in range(len(omega)-1):
    B_old=np.copy(B)
    A=K*math.sin(omega[i]*tau)-omega[i]
    B=alpha+K*math.cos(omega[i]*tau)
    tanphi[i]=A/B
    phi[i]=np.arctan(tanphi[i])
    if B*B_old<0:
        sign=-sign
    if sign<0:
        phi[i]=phi[i]-math.pi
        
delta=-np.diff(phi)/(omega[1]-omega[0])
plt.plot(omega/2/math.pi,phi)
plt.plot(omega[1:]/2/math.pi,delta)
# plt.plot(omega/2/math.pi,tanphi)
# plt.xlim(0,1)
plt.ylim(-5,5)
plt.show()

# plt.plot(omega/2/math.pi,K*np.sin(omega*tau)-omega)
# plt.plot(omega/2/math.pi,alpha+K*np.cos(omega*tau))
# plt.xlim(0.2,0.3)
# plt.ylim(-0.5,0.5)
# plt.ylim(-200,200)
# plt.grid()
# plt.show()
# print((K*np.sin(omega*tau)-omega[i]))
# plt.plot(B)
```

```python
tau_ome=np.zeros(len(omega))
for i in range(len(omega)):
    tau_ome[i]=-(K*math.cos(omega[i]*tau)-K**2*tau-alpha*(K*tau*math.cos(omega[i]*tau)-1)+K*tau*omega[i]*math.sin(omega[i]*tau))/beta[i]
plt.plot(omega/2/math.pi,tau_ome)
plt.plot(omega/2/math.pi,G)
plt.plot(omega[1:]/2/math.pi,delta)
plt.ylim(-10,10)
plt.xlim()
```
