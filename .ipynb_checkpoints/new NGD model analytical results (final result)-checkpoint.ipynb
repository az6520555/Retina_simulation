```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
alpha = 60.
beta =  2

K = 30.
g = 30.

omega=np.linspace(0,100,10000)
```

```python
A=np.zeros(len(omega))
B=np.zeros(len(omega))
for i in range(len(omega)):
    A[i]=-beta*omega[i]**2+alpha*beta**2+g*beta*K+omega[i]**2*(alpha+beta)
    B[i]=-omega[i]**3+omega[i]*alpha*beta+omega[i]*g*K-beta*omega[i]*(alpha+beta)
dA=np.diff(A)/(omega[1]-omega[0])
dB=np.diff(B)/(omega[1]-omega[0])
plt.plot(A)
plt.plot(dA)
plt.plot(B)
plt.plot(dB)
```

```python
# phase
phi=np.zeros(len(omega))
for i in range(len(omega)):
    phi[i]=math.atan(-B[i]/A[i])
plt.plot(omega/2/math.pi,phi)
```

```python
# group
delta=-np.diff(phi)/(omega[1]-omega[0])
plt.plot(omega[0:-1]/2/math.pi,delta)
plt.ylim()
```

```python
# gain
for i in range(len(omega)):
    C=(-omega[i]**2+alpha*beta+g*K)**2+(omega[i]*(alpha+beta))**2
gain=K/C*(A**2+B**2)**(1/2)
plt.plot(omega/2/math.pi,gain)
```

```python
# plot together
plt.plot(omega/2/math.pi,phi)
plt.plot(omega[0:-1]/2/math.pi,delta)
plt.plot(omega/2/math.pi,gain)
```
