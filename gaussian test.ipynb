```python
import numpy as np
import matplotlib.pyplot as plt
import math
from PIL import Image
```

```python
x=np.linspace(-1,1,100)
y=np.linspace(-1,1,100)
sigx=0.1
sigy=0.2
gx=np.exp(-x**2/sigx**2)/(sigx*np.sqrt(2*math.pi))
gy=np.exp(-y**2/sigy**2)/(sigy*np.sqrt(2*math.pi))
plt.plot(x,gx)
plt.plot(y,gy)
```

```python
g2d=np.outer(gx,gy)
X,Y=np.meshgrid(x,y)
fig=plt.figure(figsize=(10,5))
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,g2d, rstride=1, cstride=1,cmap='viridis', edgecolor='none')
```
