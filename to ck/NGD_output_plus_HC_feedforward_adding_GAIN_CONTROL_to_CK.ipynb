### generate poisson spikes

```python
# poisson as a function
# ON cell: onoff=0
# OFF cell: onoff=1
def poisson(r_in,thr,onoff):
    rtemp=(-1)**onoff*(r_in-thr)
    rtemp[rtemp<0]=0
    rtemp=rtemp/(np.sum(rtemp[100:]))*10*len(rtemp[100:]) # set the mean firing rate as 5 hz
    tempspikes=np.zeros(len(rtemp))
    for step in range(len(rtemp)):
        if random.random()<rtemp[step]*dt:
            tempspikes[step]=1
        else:
            tempspikes[step]=0
    return tempspikes
```

## NGD Adding gain control

### feedback loop from y output to x input

```python
def NGD_FBFF_GainControl(xin,h=0,psi=0.1):
    alpha,beta,gamma,K,g=6,1.6,1,22,10
    y=np.zeros(len(xin))
    z=np.zeros(len(xin))
    v=np.zeros(len(xin))
    GC=np.zeros(len(xin))
    for i in range(len(xin)-1):
        dy=dt*(-alpha*y[i]+K*(GC[i]*xin[i]-z[i]))
        dz=dt*(-beta*z[i]+g*y[i])
        dv=dt*(-gamma*v[i]+h*y[i])
        y[i+1]=y[i]+dy
        z[i+1]=z[i]+dz
        v[i+1]=v[i]+dv
        GC[i+1]=1/(1+v[i+1]**4)
    B=(1-psi)*y-psi*z
    return y,z,v,GC,B
```

### Another test of gain control

```python
def NGD_FBFF_GainControl_2(xin,h=0,psi=0.1):
    alpha,beta,gamma,kappa,K,g=6,1.6,1,1,22,10
    y=np.zeros(len(xin))
    z=np.zeros(len(xin))
    v=np.zeros(len(xin))
    GC=np.zeros(len(xin))
    B=np.zeros(len(xin))
    for i in range(len(xin)-1):
        dy=dt*(-alpha*y[i]+K*(xin[i]-z[i]))
        dz=dt*(-beta*z[i]+g*GC[i]*y[i])
        dv=dt*(-gamma*v[i]+h*z[i])
        y[i+1]=y[i]+dy
        z[i+1]=z[i]+dz
        v[i+1]=v[i]+dv
        if v[i+1]>0:
            GC[i+1]=1/(1+v[i+1]**4)
        else:
            GC[i+1]=1
        B[i+1]=(1-psi)*y[i+1]-psi*z[i+1]
    return y,z,v,GC,B
```

### gain control at two outputs

```python
def NGD_FBFF_GainControl_4(xin,h=0,f=0,psi=0.1):
    alpha,beta,gamma,kappa,K,g=6,1.6,1,0.6,22,10
    y=np.zeros(len(xin))
    z=np.zeros(len(xin))
    v=np.zeros(len(xin))
    w=np.zeros(len(xin))
    GC1=np.zeros(len(xin))
    GC2=np.zeros(len(xin))
    B=np.zeros(len(xin))
    for i in range(len(xin)-1):
        dy=dt*(-alpha*y[i]+K*(xin[i]-z[i]))
        dz=dt*(-beta*z[i]+g*y[i])
        dv=dt*(-gamma*v[i]+h*y[i])
        dw=dt*(-kappa*w[i]+f*z[i])
        y[i+1]=y[i]+dy
        z[i+1]=z[i]+dz
        v[i+1]=v[i]+dv
        w[i+1]=w[i]+dw
        if v[i+1]>0:GC1[i+1]=1/(1+v[i+1]**4)
        else:GC1[i+1]=1
        if w[i+1]>0:GC2[i+1]=1/(1+w[i+1]**4)
        else:GC2[i+1]=1
    y_gc=y*GC1
    z_gc=z*GC2
    B=(1-psi)*y_gc-psi*z_gc
    return y_gc,z_gc,GC1,GC2,B
```
