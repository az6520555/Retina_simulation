```python
import numpy as np
import matplotlib.pyplot as plt
import math 
from PyAstronomy import pyasl
from scipy import signal
%matplotlib inline
```

```python
# OU
Tot=300
dt=0.01
T=np.arange(dt,Tot,dt)
tau=0.5
D = 4
L=np.zeros(len(T))
mu, sigma = 0, 1
for i in range(len(T)-1):
    L[i+1]=L[i]*(1-dt/tau)+math.sqrt(D*dt)*np.random.normal(mu,sigma)
```

```python
plt.figure(num=1)
plt.title('OU with different cutoff frequency')
plt.plot(T,L)
fc=[10,7,4,2]
for ifc in range(len(fc)):
    b, a = signal.butter(2, 2*fc[ifc]*dt, btype='low', analog=False)
    Lf = signal.filtfilt(b, a, L)
    LfSTD = np.std(Lf)
    Lf = Lf/LfSTD
    plt.plot(T,Lf)
    plt.xlim(100,103)
```
