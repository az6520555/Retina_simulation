```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# parameter
alpha=5.3
tau=1
K=5.4
omega=np.linspace(0,5,1000)
```

```python
# phi
phi=np.zeros(len(omega))
for i in range(len(omega)):
    A=(-alpha*math.sin(-omega[i]*tau)-omega[i]+omega[i]*math.cos(-omega[i]*tau))
    B=(alpha-alpha*math.cos(-omega[i]*tau)-omega[i]*math.sin(-omega[i]*tau))
    phi[i]=math.atan(A/B)
plt.plot(omega,phi)
```

```python
# group delay
dphi=np.diff(phi)
domega=omega[1]-omega[0]
tau_omega=dphi/domega
plt.plot(omega[:-1],tau_omega)
plt.ylim(-2,0)
```

```python
# gain
G=np.zeros(len(omega))
for i in range(len(omega)):
    A=(-alpha*math.sin(-omega[i]*tau)-omega[i]+omega[i]*math.cos(-omega[i]*tau))
    B=(alpha-alpha*math.cos(-omega[i]*tau)-omega[i]*math.sin(-omega[i]*tau))
    G[i]=K/(omega[i]**2+alpha**2)*math.sqrt(A**2+B**2)
plt.plot(omega,G)
```
